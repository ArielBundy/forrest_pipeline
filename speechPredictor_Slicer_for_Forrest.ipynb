# Speech predictor slicer 

```python
import csv
```

```python
# this code opens an event tsv of speech from Forrest dataset, which is not yet compatible to our experiment,
# (tsv has narrator and word by word representations, we need sentences without narrator)
# loops over it's content and creates a text in costume 3 value column for FSL use.
# note that for our experiment we need to use 2 segments (7+8)!

# we first load the tsv file and read line by line to identify rows we need (not narrator, whole sentences)
# and collect them to 'lines' (a list)
tsv_file = open("speech_ad_run-8_events.tsv")
read_tsv = csv.reader(tsv_file, delimiter="\t")
lines = [];
for row in read_tsv:
    if row[4] =='SENTENCE' and row[2] !='NARRATOR':
    # collect all important data: time, duration, speaker, sentence.    
        lines.append([row[0],row[1],row[2],row[3]]) 
```

```python
# uncomment for validation 
#for line in lines:
#    print(line)
```

```python
# create a file with all lines
with open("Forrest_sentences-seg8.csv", "w") as output_file:    
    writer = csv.writer(output_file,delimiter = '\t')
    writer.writerows(lines)
```

```python
# after creating both segments and manualy editing them into one file with only events we will use
# we will correct the timeline to start at 0

# segment 7
edit_file = open("Forrest_sentences-seg7_edit.csv","r")
read_csv = csv.reader(edit_file, delimiter="\t")
# 'corrected_lines' is our predictor (costume 3 column for FSL)
corrected_lines=[]
for line in read_csv:
    line[0] = float(line[0])
    line[0] -= 806.4 # based approximation for segment 7
    line[0] = float("{:.3f}".format(line[0]))
    corrected_lines.append([line[0],line[1],1]) # taking first two values (time and duration) adding 1 for 3rd column
    last = line[0]
    print(line) # for validation
```

```python
# segment 8
# we need to connect the end of segment 7, seconds from 'Ernie and Bert' until 'Forrest' were counted and corrected
edit_file = open("Forrest_sentences-seg8_edit.csv","r")
read_csv = csv.reader(edit_file, delimiter="\t")
for line in read_csv:
    line[0] = float(line[0])
    line[0] += last + 10
    line[0] = float("{:.3f}".format(line[0]))
    print(line) # for validation
    corrected_lines.append([line[0],line[1],1])
    
```

```python
with open("Forrest_speech_predictor.txt", "w") as output_file:    
    writer = csv.writer(output_file,delimiter = '\t')
    writer.writerows(corrected_lines)
```
